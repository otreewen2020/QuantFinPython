## Estimation of parameters of Geometric Brownian Motion

The script below uses data available at Stooq to find parameters for the GBM process that can be used to model the evolution of asstets prices
NOTE: obviously it does not mean that the asset priced actually follow the GBM, because they do not - 
cf. below

```python
# imports
import pandas as pd
import get_data_from_stooq
import numpy as np
import scipy.stats as spst
import math
import matplotlib.pyplot as plt
%matplotlib notebook
import statsmodels.api
```

```python
# prepare list of ticker symbols to analyze
list_of_tickers = ["tpe", "kgh", "peo", "lwb", "pkn", "lts", "opl"]
```

```python
# get data from stooq
ts_start_date = "2011-01-01"
ts_end_date = "2016-12-31"
data_df = get_data_from_stooq.get_daily_data_from_stooq_for_tickers(list_of_tickers,
                                                                   ts_start_date,
                                                                   ts_end_date)
```

```python
# printing a comparative plot of relative price changes
data_df_normd = data_df/data_df.iloc[0,:]
print("Head of normed data: ")
print(data_df_normd.head(3))
print("Tail of normed data: ")
print(data_df_normd.tail(3))
data_df_normd.plot()
plt.grid(True)
plt.title("Comparison of standardized performance over the analyzed period")
plt.xlabel("Time")
plt.ylabel("Standardized value/cumulative return")
```
